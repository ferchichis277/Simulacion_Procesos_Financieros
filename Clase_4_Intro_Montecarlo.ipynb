```python
import numpy as np 
import matplotlib.pyplot as plt
```

### Calculadora de integrales definidas:

```python
def funcion_a_integrar(x):
    return x**2
a = 2
b = 5 

N = 1000000

x = np.random.uniform(a,b,N)
y = np.random.uniform(0, funcion_a_integrar(b), N)
```

```python
n_bajo_curva = np.sum(y < funcion_a_integrar(x))
n_bajo_curva
```

```python
respuesta = (b-a) * funcion_a_integrar(b) * n_bajo_curva/N
respuesta
```

```python
plt.plot(x,y,'.')
plt.plot(np.linspace(a,b,100), funcion_a_integrar(np.linspace(a,b,100)), 'r')
plt.show()
```

## Actividad 1: Calcular Pi

```python
# Definir puntitos
N = 100000

# Definir l√≠mites del cuadrado
lado_min = -1
lado_max = 1
```

```python
# Generar las coordenadas aleatorias
x = np.random.uniform(lado_min, lado_max, N)
y = np.random.uniform(lado_min, lado_max, N)
```

```python
# Puntos que caen dentro del circulo
distancia_al_centro = x**2 + y**2
puntos_dentro_circulo = np.sum(distancia_al_centro <= 1)
```

```python
# Calcular pi
pi_estimado = 4 * (puntos_dentro_circulo / N)
pi_estimado
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y, '.')
theta = np.linspace(0, 2*pi_estimado, 100)
plt.plot(np.cos(theta), np.sin(theta), 'r', linewidth=2)
plt.axis('equal') # Para que el c√≠rculo se vea redondo y no ovalado
plt.show()
```
